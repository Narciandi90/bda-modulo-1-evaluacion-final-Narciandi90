A lo largo del ejercicio tendré que crear una clase llamada TiendaOnline con una serie de requisitos en cuanto a sus atributos.
El primero de ellos se denominará **inventario** en el que almacenar cada producto representado por una lista de diccionario:
**inventario** = [{"nombre":, "precio":, "cantidad"}, {"nombre":, "precio":, "cantidad"}]
El segundo de los atributos se llamará **ventas_totales** y llevará un registro de las ventas totales de la tienda.

```python
inventario = [{"nombre": "precio": "cantidad":}, {"nombre": "precio": "cantidad":}]
```

```python
'''Comencemos creando la clase tal y como se nos indica y desarrollando 
los distintos métodos de la misma'''

class TiendaOnline:
    def __init__(self): #El primer paso es el método constructor
        self.inventario = [] #Lista inicialmente vacía
        self.ventas_totales = 0.0 #Atributo tipo float que se irá actualizando
        self.clientes = {}

    def agregar_productos(self, nombre, precio, cantidad): #Definimos el primero de los métodos
       for i in self.inventario: #Cada elemento de la lista inventario (i) es un diccionario
           if nombre == i["nombre"]: #Por eso debemos acceder a ellos por clave
               print(f"El producto {nombre} ya está registrado")
               i["cantidad"] += cantidad
               print(f"La cantidad de {nombre} ha sido actualizada")
               return
       nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}
       self.inventario.append(nuevo_producto)#self.inventario es una lista
       print(f"El producto {nombre} ha sido agregado exitosamente")
    
    def ver_inventario(self):
        for i in self.inventario: #Accedemos a  los elementos de inventario. Recordemos que cada uno es un diccionario
            print(f"Nombre: {i["nombre"]}, Precio: {i["precio"]}€, Cantidad: {i["cantidad"]}")
            
    def buscar_producto(self, nombre):
        for i in self.inventario:
            if nombre == i["nombre"]:
                print("Los datos del producto que está buscando son los siguientes:")
                print(f"Nombre: {nombre}, Precio: {i["precio"]}€, Cantidad: {i["cantidad"]}")
                return
        print(f"El producto {nombre} no está disponible")
    
    def actualizar_stock(self, nombre, cantidad):
        for i in self.inventario:
            if nombre == i["nombre"]:
                i["cantidad"] += cantidad
                print(f"El stock de {nombre} ha sido actualizado")
                print(f"Las unidades disponibles son {i["cantidad"]}")#Mostrar la cantidad actualizada
                return
        print(f"{nombre} no se encuentra en nuestro inventario")
            
    def eliminar_producto(self, nombre):
        for i in self.inventario:
            if nombre == i["nombre"]:
                self.inventario.remove(i) #Usamos remove porque el elemento i pertenece a la lista inventario
                print(f"{nombre} ha sido eliminado del inventario con éxito")
                return
        print(f"El producto {nombre} no se encuentra en el inventario")
    
    def calcular_valor_inventario(self):
        valor_inventario = 0 #Necesitamos una variable("una hucha") 
        for i in self.inventario:
            valor_inventario += i["precio"]*i["cantidad"]
        return print(f"El valor total de las exsitencias es de {valor_inventario} €")
    
    def agregar_cliente(self, nombre, email):
        if nombre in self.clientes: 
            print("Cliente ya registrado")
            return
        else:
            self.clientes[nombre] = {"email":email, "compras":[]}#Si el cliente no existe creo la clave con su nombre 
            print(f"El nuevo cliente {nombre} ha sido registrado")

    def ver_clientes(self):
        for nombre, datos_detalle in self.clientes.items():
            email_cliente = datos_detalle["email"]
            print(f"Nombre del cliente: {nombre}\n Correo electrónico: {email_cliente}")
            print("-"*10)
         
         

            
        
           
            
                         
        
      
       
                                    
              

        
```
